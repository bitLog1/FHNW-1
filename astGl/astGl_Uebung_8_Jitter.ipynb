### Aufgabe

```python
sig_jps =   1.0E-12 # 1psRMS
sig_jfs = 100.0E-15 # 100fsRMS
```

```python
f = np.logspace(1, 8)
SJNRps = 1/(8*np.pi**2 * f**2 * sig_jps**2)
SJNRps_db = 20*np.log10(SJNRps)
SJNRfs = 1/(8*np.pi**2 * f**2 * sig_jfs**2)
SJNRfs_db = 20*np.log10(SJNRfs)
```

```python
plt.figure(0)
plt.plot(f, SJNRps_db)
plt.plot(f, SJNRfs_db)
plt.legend(['sigma_j = 1ps', 'sigma_j = 100fs'])
plt.yscale('linear')
plt.xscale('log')
plt.grid(True)
plt.gca().xaxis.grid(True, which='minor')  # minor grid on too
plt.xlabel('f [Hz]')
plt.ylabel('SJNR [dB]')
```
